```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import warnings
warnings.filterwarnings('ignore')
```

### inciso 1
Genere un gráfico que indique el número de medallas concedidas en cada uno de los cinco principales deportes en 2016.

### Solución



Para realizar la actividad, realizaremos los siguientes pasos:

- paso 1:
 Descargue el conjunto de datos y formatéelo como un DataFrame de pandas.

```python
medallas = pd.read_csv('C:/Users/Usuario/Downloads/athlete_events.csv')
medallas.head(5)
```

- paso 2:
 Filtra el DataFrame para incluir solo las filas correspondientes a los ganadores de medallas de 2016.
- paso 3:
Descubre las medallas concedidas en 2016 en cada deporte.

```python
medalsnew = medallas[medallas['Medal'].notna()]
medconceb = medalsnew[medalsnew['Year'] == 2016]
medconceb['Sport'].value_counts()
```

Podemos observar que, los 5 deportes con más medallas son:
 - Athletics
 - Swimming
 - Rowing
 - Football
 - Hockey

- paso 4:
 Enumera los cinco deportes más importantes en función del mayor número de medallas concedidas. Filtra el DataFrame una vez más para incluir solo los registros de los cinco deportes principales en 2016.

```python
medalsnew = medallas[medallas['Medal'].notna()]
medconceb = medalsnew[medalsnew['Year'] == 2016]
countmedals = medconceb['Sport'].value_counts()
topfive = countmedals.head(5)


topsports = ['Athletics', 'Swimming', 'Rowing', 'Football', 'Hockey']

newtop = medconceb[medconceb['Sport'].isin(topsports)]
newtop.head()
```

- paso 5:
 Genere un gráfico de barras con los recuentos de registros correspondientes a cada uno de los cinco deportes principales.

```python
sns.set_style("dark")
sns.barplot(x=topfive.index, y=topfive.values, palette="magma", edgecolor='black')
plt.title('Medallas concebidas al top 5 deportes')
plt.xlabel('Deportes')
plt.ylabel('Medallas')
plt.show()
```

### Inciso 2
Trace un gráfico que represente la distribución de la edad de los ganadores de medallas en los cinco principales deportes en 2016.


- paso 6:
 Generar un histograma para la característica Edad de todos los ganadores de medallas en los cinco deportes principales (2016).

```python
topages = newtop['Age']
color = cm.magma(0.5)
sns.histplot(topages, color=color, edgecolor = 'black')
plt.title('Edades de los ganadores de medallas')
plt.xlabel('Edades')
plt.ylabel('Cantidad')
plt.show()
```

De la gráfica anterior podemos concluir que las edades de los jugadores se distribuyen entre 20 y 32 años aproximadamente.

### Inciso 3
Descubre qué equipos nacionales ganaron el mayor número de medallas en los cinco principales deportes en 2016.

- paso 7:
 Genera un gráfico de barras que indique cuántas medallas ganó el equipo de cada país en los cinco deportes principales en 2016.

```python
teamstable = pd.crosstab(index = newtop['Team'], columns = 'count')
newtable = teamstable.sort_values(by='count', ascending=False)
newtable.plot(kind='bar',color="#8B008B",legend=False)
plt.show()

```

Los 3 equipos de los paises que ganaron mas medallas fueron: Estados Unidos, Alemania y Gran Bretaña.

### Inciso 4
Observe la tendencia del peso medio de los atletas masculinos y femeninos ganadores en los cinco principales deportes en 2016

- paso 8: 
Genere un gráfico de barras que indique el peso medio de los jugadores, clasificados en función del género, que ganaron en los cinco principales deportes en 2016.



```python
pesoprom = newtop.groupby('Sex')['Weight'].mean()
sns.set_style("dark")
pesoprom.plot(kind='bar',color="#8B008B",legend=True)
plt.title('Medallas concebidas al top 5 deportes según el género')
plt.xlabel('Género')
plt.ylabel('Medallas')
plt.show()
```

Se puede observar que las mujeres en el top 5 de deportes con más medallas ganadas tienen menor peso que los hombres.
